```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import ssl
```

```python
#!/usr/bin/env python3.10

# fixing computer specific issue when downloading training material
ssl._create_default_https_context = ssl._create_unverified_context

# loading the data and establishing y/x_train and x/y_test
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

#helps set the data and pixels
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)

# creating keras api neural network for classifing handwritten images
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten(input_shape=(28, 28)))
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))

# configure the optimizer and the loss for classifcation
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Load the previously saved model if available
try:
    prev_model = tf.keras.models.load_model('Handwritten.model')
    prev_accuracy = prev_model.evaluate(x_test, y_test, verbose=0)[1]
except:
    prev_accuracy = 0.0  # If no previous model is available, start with 0 accuracy

# Train the model
history = model.fit(x_train, y_train, epochs=400)

# Evaluate the accuracy of the newly trained model
new_accuracy = model.evaluate(x_test, y_test, verbose=0)[1]

# Save the new model only if it has higher accuracy than the previous model
if new_accuracy > prev_accuracy:
    model.save('Handwritten.model')
    print("New model saved due to higher accuracy:", new_accuracy)
else:
    print("New model not saved due to lower accuracy:", new_accuracy)
```

```python
from PIL import Image

# Load the saved model
model = tf.keras.models.load_model('{ENTER FILE PATH}/Handwritten.model')

# Load and preprocess the image
def preprocess_image(image_path):
    image = Image.open(image_path).convert('L')
    image = image.resize((28, 28))
    image_array = np.array(image)
    normalized_image = image_array / 255.0
    return tf.keras.utils.normalize(normalized_image, axis=1)

image_path = '{ENTER FILE PATH}/number.png'
input_image = preprocess_image(image_path)
reshaped_input = input_image.reshape(1, 28, 28)  # Add batch dimension for prediction

# Make predictions
predictions = model.predict(reshaped_input)

# Interpret predictions
predicted_class = np.argmax(predictions)

# Display prediction results
# Class is the number and the bigger the score of a class determines the confidence

print("Predicted class:", predicted_class)
print("Raw prediction scores:", predictions)
```
